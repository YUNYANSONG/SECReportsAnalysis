```python
import requests
from bs4 import BeautifulSoup
import os
import numpy as np
import pandas as pd
from statsmodels.regression.linear_model import OLS

# Define URL of the website you want to scrape
#website_url = 'https://www.citadelsecurities.com/rule-605-606-statements/'
website_url = "https://www.virtu.com/about/transparency/rule-605-and-606-reporting/"
website_url = "https://clearingcustody.fidelity.com/trade-execution-quality/sec-rule-606"
# Make an HTTP request to fetch the raw HTML content
response = requests.get(website_url)
if response.status_code != 200:
    print(f"Failed to get webpage: {response.status_code}")
    exit(1)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Define directory to save files
save_directory = "downloaded_reports"
os.makedirs(save_directory, exist_ok=True)

company_name_list = ["TVIRTU","TCDRG",]

# Loop through the links and download Rule 605 and 606 reports
# Assuming the reports have the keywords '605' or '606' in their names or URLs
for link in soup.find_all('a'):
    file_url = link.get('href')
    if file_url and ('TVIRTU' in file_url or 'xml' in file_url):
        print(f"Downloading: {file_url}")
        
        # Download the file
        r = requests.get(file_url)
        if r.status_code == 200:
            file_name = os.path.join(save_directory, file_url.split('/')[-1])
            if not os.path.exists(file_name):
                with open(file_name, 'wb') as f:
                    f.write(r.content)
        else:
            print(f"Failed to download file: {r.status_code}")
```
